### 화염 합성법을 이용한 삼원계 양극재의 자원, 에너지 소비 및 경제성 분석

**Wen DU, Junlei WANG, Yunfei XU, Shilong LI, Kun WANG**  
**개요 | 0 공유 | 39 인용**  
**https://doi.org/10.1016/j.jclepro.2024.143499**  
**전문 접근 가능**

#### 초록
양극재는 리튬이온 배터리 제조 비용의 약 3분의 1을 차지하며, 이는 전기차용 배터리 팩의 더 넓은 적용을 막는 주요 요인입니다. 화염 합성법(FS)은 장비가 적고 효율성이 높은 나노구조 물질의 새로운 제조 기술로, 양극재 제조 비용을 크게 낮출 가능성이 있습니다. 이 연구의 주된 목표는 화염 합성법을 통해 대규모 양극재 생산 시 자원, 에너지 소비, 그리고 경제적 타당성을 정량적으로 평가하는 것입니다. 분석 과정에서 반응물과 생성물의 질량 흐름, 그리고 화염 합성법으로 생산된 삼원계 양극재 $\mathrm{LiNi}_x \mathrm{Co}_y \mathrm{Mn}_{(1-x-y)} \mathrm{O}_2$ (단, $x+y+z=1$)의 최소 판매 가격(MCSP)을 계산하여 기존의 수산화물 공침법과 비교했습니다. 결과에 따르면, 화염 합성법으로 삼원계 양극재를 생산하면 $\mathrm{CO}_2$ 배출량이 약 60%, 물 소비량이 약 30% 감소합니다. $\mathrm{LiNi}_{0.5} \mathrm{Co}_{0.2} \mathrm{Mn}_{0.3} \mathrm{O}_2$ (NCM523), $\mathrm{LiNi}_{0.6} \mathrm{Co}_{0.2} \mathrm{Mn}_{0.2} \mathrm{O}_2$ (NCM622), $\mathrm{LiNi}_{0.8} \mathrm{Co}_{0.1} \mathrm{Mn}_{0.1} \mathrm{O}_2$ (NCM811)의 MCSP는 각각 224.8천 위안/톤, 235.1천 위안/톤, 240.0천 위안/톤으로, 중국 시장 평균 판매가보다 약 2%, 10%, 14% 낮아 경제적 타당성을 보여줍니다. 특히 고니켈 양극재 생산에서 화염 합성법은 생산 공정을 단순화하고 비용을 줄이는 데 큰 잠재력을 보였습니다. NCM811에 대한 민감도 분석 결과, 원료 가격, 연료 및 전력 비용이 25% 감소하면 MCSP가 186.4천 위안/톤까지 내려갈 수 있습니다. 용액의 금속 양이온 농도가 3.2몰/L로 증가하면 중국의 가장 엄격한 에너지 소비 제한 기준을 충족할 수 있습니다. 리튬 과잉분을 10%에서 5%로 줄이면 NCM811의 MCSP가 추가로 5.6천 위안/톤 감소하며, 이는 화염 합성법의 고온으로 인한 리튬 손실이 제조 비용 감소에 적당한 영향을 미친다는 점을 보여줍니다.

### 1. 서론
리튬이온 전원 배터리는 높은 에너지 밀도, 긴 수명, 낮은 자가 방전율, 높은 에너지 활용 효율 등의 장점 덕분에 전기차와 하이브리드 전기차에서 넓은 응용 가능성을 가집니다(Zhang et al., 2024). 그러나 교통의 빠른 전기화는 전기차 공급망을 혼란스럽게 하고 배터리 산업 전반에 걸쳐 가격 변동을 일으키고 있습니다(Yang et al., 2024). 리튬이온 배터리의 가격은 주로 양극재에 의해 결정되며, 이는 제조 비용의 약 30~35%를 차지하고, 생산 조건, 원료 가격, 수요-공급 관계 등에 영향을 받습니다(Wood et al., 2015). $\mathrm{LiNi}_x \mathrm{Co}_y \mathrm{Mn}_{(1-x-y)} \mathrm{O}_2$ (NCM) 삼원계 양극재는 뛰어난 에너지 밀도, 속도 성능, 저온 저항성으로 인해 글로벌 전원 배터리 시장을 지배하며, 세계 시장 점유율의 60%를 차지합니다(IEA, 2023). 현재 업계는 주로 수산화물 전구체를 공침법으로 준비한 후 리튬을 혼합하고 소결하여 양극재를 생산합니다(Han et al., 2022).

## <페이지 4>
### <OCR에서 추출된 내용>
공침법은 여러 합성 단계를 거치며, 긴 생산 주기, 까다로운 정제 과정, 폐액 부산물 생성, 소결 단계에서의 높은 합성 온도와 탄소 배출 등의 단점이 있습니다. 특히 고니켈 삼원계 양극재 생산에서는 여러 차례의 긴 열처리가 필요해 합성 시간이 길어지고 에너지 소비가 증가하는 문제점이 있습니다(Bai et al., 2020; Li et al., 2016; Xiong et al., 2013; Xu et al., 2019).

화염 합성법(FS)은 나노구조 물질을 제조하는 새로운 기술입니다. FS의 주요 과정은 전구체 용액을 마이크론 크기의 액적 형태로 분무한 후, 화염 연소로 생성된 고온 환경에 도입하는 것입니다. FS는 높은 온도, 큰 열 구배, 매우 짧은 고온 체류 시간 덕분에 불순물이 적고 균일한 크기의 입자를 단일 단계로 합성할 수 있습니다(Pratsinis, 1998). FS의 역사를 돌이켜보면, 이 방법은 선사 시대 동굴 벽화와 중국 먹물 예술에서 사용된 이래로 오랜 역사를 가집니다(Debecker et al., 2018; Pratsinis, 1998). 1971년 Ulrich는 층류 예혼합 화염에서 $\mathrm{SiCl}_4$를 사용해 $\mathrm{SiO}_2$ 분말을 합성하며 기체 전구체를 이용한 FS를 개척했습니다(Ulrich, 1971). 1977년 Sokolowski 등은 알루미늄 아세틸아세톤 액적을 벤젠-에탄올에서 연소시켜 $\mathrm{Al}_2 \mathrm{O}_3$ 나노입자를 제조하는 방법을 처음 제안했습니다(Sokolowski et al., 1977). 1996년과 1997년 Bickmore 등과 Karthikeyan 등은 가스 보조 분무기 화염 합성 반응기를 통해 체계적인 FS 연구를 시작했습니다(Bickmore et al., 1996; Karthikeyan et al., 1997). 2002년 이후 Pratsinis와 스위스 연방 공과대학(취리히)의 Baiker, Madler, Wegner 등은 단일 금속 산화물과 복합 금속 산화물에 대한 체계적인 연구를 이끌었습니다(Pratsinis, 1998, 2010; Teoh et al., 2010).

산업 응용 가능성과 공정 제어 측면에서, FS 반응기의 확장 연구는 Mueller 등(2003, 2004)과 Gröhn 등(2012, 2014)에 의해 연속적으로 수행되었습니다. 결과는 FS의 공정 매개변수를 쉽게 제어할 수 있음을 보여주며, 산업 장비가 실험실 조건과 유사한 물질 준비를 달성할 기술적 타당성을 확립했습니다. FS는 최근 수십 년간 빠르게 발전하며 촉매, 센서, 배터리, 치과, 생의학 등 다양한 분야에서 유망한 응용 가능성을 보여주었습니다(Strobel and Pratsinis, 2007). 카본 블랙(Cabot, Columbia, Degussa-Hüls), 흄드 실리카(Cabot, Degussa-Hüls), 안료 티타니아(DuPont, Ishihara, Millennium, Kerr-McGee, Cristal), 광섬유(Corning, Heraeus, Lucent, Sumitomo)와 같은 일부 상품 나노입자 제품은 FS로 제조되며, 연간 수백만 톤 생산으로 150억 달러 이상의 가치를 창출합니다(Teoh et al., 2010). 리튬이온 배터리의 양극재 또는 음극재 합성을 위한 FS는 최근 몇 년간 주요 연구 분야로 떠올랐습니다. 양극재 연구는 주로 삼원계 양극재(Kim et al., 2012; Wang et al., 2018), $\mathrm{LiCoO}_2$(Choi et al., 2013; Hussaini et al., 2023; Jang et al., 2004), $\mathrm{LiMn}_2 \mathrm{O}_4$(Ernst et al., 2007; Patey et al., 2009; Saadatkhah et al., 2019), $\mathrm{LiFePO}_4$(LFP)(Oliver Waser, 2010; Pratheeksha et al., 2016; Waser et al., 2011)에 초점을 맞췄습니다. 삼원계 양극재의 경우, Yamada 등(2009)은 높은 수율의 FS 반응기를 개발하여 $\mathrm{LiNi}_{1/3} \mathrm{Co}_{1/3} \mathrm{Mn}_{1/3} \mathrm{O}_2$ (NCM333)를 250시간 동안 시간당 2kg씩 연속 생산할 수 있었습니다. Ogihara 등(2009)의 연구는 파일럿 장치(월 2000kg 생산 능력)로 얻은 양극재의 전기화학적 특성이 동일 조건의 실험실 장치와 일치함을 보여주었습니다. Wang 등(2018)은 화염 분무 열분해를 통해 NCM333을 단일 단계로 합성했으며, 전기화학 결과는 높은 전구체 농도가 안정적인 사이클 수명을 제공함을 나타냈습니다.

Zhang 등은 바이오매스 글리세롤을 연료로 사용한 FS 방법을 개발하여 NCM333(Zhang et al., 2020a)과 $\mathrm{LiNi}_{0.8} \mathrm{Co}_{0.15} \mathrm{Al}_{0.05} \mathrm{O}_2$ (NCA)(Zhang et al., 2020b)를 성공적으로 합성했습니다. Liang 등(2021)은 실험실 규모의 FS 플랫폼을 활용해 고니켈 삼원계 양극재 $\mathrm{LiNi}_{0.6} \mathrm{Co}_{0.2} \mathrm{Mn}_{0.2} \mathrm{O}_2$ (NCM622)와 $\mathrm{LiNi}_{0.8} \mathrm{Co}_{0.1} \mathrm{Mn}_{0.1} \mathrm{O}_2$ (NCM811)를 하루 1.5kg 속도로 합성했습니다. Abram 등(2019)은 FS로 NCM811을 제조하여 84.3%의 쿨롱 효율과 181 mAh/g의 초기 방전 용량을 달성했습니다. Zang 등(2021)의 연구는 FS로 NCM333을 생산하는 기술-경제 분석을 수행하여, MCSP가 탄산염 공침법보다 17% 낮음을 밝혔습니다. 이러한 결과는 FS를 이용한 삼원계 양극재 생산이 공정 타당성을 보이며 비용 절감에 큰 잠재력을 가짐을 보여줍니다.

높은 생산 효율과 간단한 장비로 FS는 기존 방법이 직면한 비용, 효율성, 환경 문제를 해결할 것으로 기대됩니다. 다양한 방법으로 삼원계 재료를 합성하는 데 소요되는 시간 비교는 표 1에 나와 있습니다. 고상 및 액상 방법과 달리 FS는 삼원계 양극재 준비 시간을 한 자리 수 수준으로 크게 줄입니다(Chen et al., 2024). 실험 결과, FS로 제조된 NCM 재료는 공침법으로 만든 것보다 입자 크기가 작습니다(Abram et al., 2019; Yan et al., 2021; Zhang et al., 2022; Zhang, J. et al., 2023). 전기화학 성능 테스트에서 FS로 제조된 NCM 재료는 공침법과 비슷한 성능을 보이며, 자세한 비교는 표 S1에 있습니다.

혼합 단계에서는 질산염을 순수한 물과 Li:Ni:Co:Mn=1.1:x:y:z의 몰 비율로 혼합기에 넣습니다. 용액은 용액 탱크에서 4시간 동안 교반된 뒤 80°C로 가열되며(Zhang et al., 2020a), 이후 전구체 용액에 2.5wt%의 요소($\mathrm{CO}\left(\mathrm{NH}_2\right)_2$)가 추가됩니다. 화염 합성 단계에서는 전구체 용액이 회전 분무기로 분무되어 마이크론 크기의 액적이 FS 반응기에서 생성된 고온 환경에 도입됩니다. 이 강렬한 환경에서 액적의 물이 증발하며 핵 생성, 충돌, 융합 등의 물리화학 반응이 일어나 나노입자가 형성됩니다. 생성된 나노입자는 사이클론 수집기로 모아집니다. NCM 재료 합성에서 요소의 영향은 이전 연구에서 조사되었으며(Zhang et al., 2022), 요소 추가는 (003) 및 (101) 피크의 열적 이동을 크게 완화하고, 소결 조건에 대한 결정화 및 양이온 혼합의 민감도를 줄이는 것으로 나타났습니다. 또한, 20분 소결로 공침법과 유사한 층상 구조와 전기화학적 특성을 얻을 수 있었으며, 본 연구에서는 제품 품질 보장을 전제로 소결 시간을 2시간으로 설정했습니다. FS 반응기는 천연가스로 연료를 공급받고, 소결 단계는 전기 에너지를 사용합니다. 소결 후 양극재는 제트 밀에서 분쇄되며, 이후 분말은 배치, 제철, 포장 과정을 거칩니다.

주요 장비 외에도 사일로, 밸브, 계량 시스템, 먼지 수집 시스템, 생산 라인, 공압 운반 시스템 등의 보조 설비가 필요합니다. 질산염 사용으로 질소 산화물($\mathrm{NO}_x$)이 발생하므로, 습식 스크러버(Flagiello et al., 2022)와 알칼리 물질, 선택적 촉매 환원(SCR) 탈질 장치를 사용해 $\mathrm{NO}_x$ 배출을 처리하며, SCR 입구 온도는 250°C~300°C로 안정화됩니다(Wang et al., 2021). SCR 처리 후 배기가스는 암모니아 흡수탑에서 잉여 암모니아를 흡수합니다. 또한, 백 필터 장치를 통해 먼지를 수집합니다. FS 공정 개선 가능성을 종합적으로 평가하기 위해 열교환기(Ding et al., 2024)를 FS 시스템에 설치하는 시나리오도 고려했습니다. 처리 후 배기가스는 전구체 용액 예열을 위한 작동 유체로 사용되며, 공정 및 열 회수 효과 평가는 보충 자료(SM) 섹션 4에 자세히 나와 있습니다.

(2) **수산화물 공침 공정**  
FS와 비교하기 위해 수산화물 공침법을 사용했으며, 동일한 생산 규모를 유지했습니다. 수산화물 공침법의 개요 흐름은 그림 3에 나와 있습니다. 황산니켈 육수화물($\mathrm{NiSO}_4 \cdot 6 \mathrm{H}_2 \mathrm{O}$), 황산코발트 칠수화물($\mathrm{CoSO}_4 \cdot 7 \mathrm{H}_2 \mathrm{O}$), 황산망간 일수화물($\mathrm{MnSO}_4 \cdot \mathrm{H}_2 \mathrm{O}$)을 순수한 물과 Ni:Co:Mn=x:y:z 몰 비율로 첫 번째 혼합기에 넣어 황산염 용액을 준비합니다. 수산화나트륨($\mathrm{NaOH}$) 용액은 두 번째 혼합기에서 준비됩니다. 연속 교반 탱크 반응기(CSTR)에서 두 용액은 10시간 동안 연속 교반됩니다(Ahmed et al., 2017). 침전 단계 후, 슬러리는 분리기에서 액체-고체 분리를 거치고, 습윤 분말은 필터에서 여러 번 여과된 후 폐액이 배출됩니다. 건조 후, 수산화물 분말은 볼 밀에서 목표 입자 크기로 분쇄되고, 탄산리튬($\mathrm{Li}_2 \mathrm{CO}_3$) 분말과 혼합되어 소결 kiln으로 보내집니다. 니켈 함량 증가로 $\mathrm{Ni}^{2+}$ 형성과 양극 표면의 잔류 리튬이 발생하므로, 고니켈 삼원계 양극재의 소결 단계는 수산화리튬 일수화물($\mathrm{LiOH} \cdot \mathrm{H}_2 \mathrm{O}$)을 사용해 낮은 소결 온도를 적용합니다. Ni가 60% 미만이면 한 번의 소결로 잔류 리튬을 허용 범위 내로 제어할 수 있지만, 60% 이상이면 추가 처리와 여러 번의 소결이 필요합니다(Bai et al., 2020).

##### 2.2.2. 질량 균형 계산
질량 균형 계산의 목적은 원료 요구량과 제품을 결정하는 것입니다. 원료는 질산염 복합체, 요소, 순수한 물입니다. 질산염 용액의 일일 할당량은 금속 양이온 농도와 일일 생산 능력을 기반으로 식 (1)을 통해 계산됩니다.

$$
V_S = \frac{\sum \frac{n_{\text {mel }}}{p_{\text {mel }}}}{\left(1 - r_M\right) \cdot M_{\text {NCM}} \cdot c} \times 10^3
$$

여기서 $V_S$는 전구체 용액의 일일 할당량($\mathrm{m}^3/\mathrm{d}$), $n_{\text {mel }}$은 Li, Ni, Co, Mn 질산염의 화학량론적 수, $p_{\text {mel }}$은 Li, Ni, Co, Mn 질산염 순도(%), $W$는 FS 공장의 생산 능력(t/d), $r_M$은 제조 중 제품 손실률(%), $M_{\text {NCM}}$은 삼원계 양극재의 몰 질량(g/mol), $c$는 전구체 용액의 설정 농도(mol/L)입니다. 일일 질산염 용액 할당량에서 각 고체 질산염의 수요는 식 (2)로 구할 수 있습니다.

$$
m_{\text {me,i}} = \frac{M_{\text {me,i}} \cdot V_S \cdot c}{\omega_{\text {me,i}}} \times \frac{n_{\text {me,i}} / p_{\text {me,i}}}{\sum n_{\text {me,i}} / p_{\text {me,i}}} \times 10^{-3}
$$

여기서 $m_{\text {me,i}}$는 Li, Ni, Co, Mn 질산염의 일일 수요(t/d), $M_{\text {me,i}}$는 Li, Ni, Co, Mn의 몰 질량(g/mol), $\omega_{\text {me,i}}$는 질산염에서 Li, Ni, Co, Mn의 이론적 금속 질량 분율입니다. 요소 요구량은 총 용액 질량의 2.5%로 계산됩니다.

질산염이 용액 준비 중 결정수를 전구체 용액에 대체하므로, 준비 중 수분 평형 계산이 필요합니다. 계산을 단순화하기 위해 혼합 단계에서 다양한 수용액의 부피가 일정하다고 가정했습니다. 각 질산염이 가져오는 물의 양은 질산염 질량에 물 함량을 곱해 구하고, 질산염 용액 할당에 추가로 필요한 순수한 물은 식 (3)으로 계산됩니다.

$$
V_{PW} = V_S - V_{CW} = V_S - \frac{\sum \left( m_{\text {me,i}} \omega_{\text {me,water}} p_{\text {me,i}} \right)}{\rho_{\text {water}}}
$$

여기서 $V_{PW}$는 염 용액에 추가된 순수한 물의 양($\mathrm{m}^3/\mathrm{d}$), $V_{CW}$는 용액에 포함된 결정수의 양($\mathrm{m}^3/\mathrm{d}$), $\omega_{\text {me,water}}$는 각 Li, Ni, Co, Mn 질산염에 포함된 결정수의 질량 분율, $\rho_{\text {water}}$는 물의 밀도($\mathrm{kg}/\mathrm{m}^3$)입니다.

질량 균형 계산에서 용액 농도는 2mol/L, 생산 과정 중 손실률은 2%, 질산염 순도는 배터리 등급 리튬, 니켈, 코발트, 망간 원료에 대한 산업 요구 한계 내로 가정했습니다. FS 원료의 물리적 정보는 표 2에 나와 있습니다. 수산화물 공침법으로 생산하는 원료의 물리적 매개변수는 보충 자료(SM)의 표 S2를 참조하세요.

공장 전체 생애주기에 대한 연간 현금 흐름은 총 자본 투자(TCI)와 연간 운영 비용(AOC)(표 6 참조)을 기반으로 계산되었습니다. 할인된 현금 흐름 방법을 사용하여 재료의 최소 양극재 판매 가격(MCSP)을 결정했으며, 순 현재 가치(NPV)가 0이 되도록 했습니다. 이 지표는 양극재 제조 비용 절감 효과를 평가하는 데 사용되었습니다. NPV는 식 (6)을 통해 계산됩니다.

$$
NPV = \sum_{t=0}^n \frac{\left(CF_I - CF_O\right)_t}{\left(1 + i_e\right)^t}
$$

여기서 $t$는 공장의 생산 및 운영 기간, $i_c$는 할인율(%), $\left(CF_I - CF_O\right)_t$는 기간 $t$ 동안의 순 현금 흐름입니다.

내부 수익률(IRR)은 국가 발전 개혁 위원회(NDRC)와 주택 도시 농촌 개발부(MOHURD) 데이터(NDRC and MOHURD, 2013)에 따라 신규 화학 재료에 대한 세후 최소 매력 수익률(MARR)과 비교되었으며, FS의 경제적 타당성을 더 포괄적으로 보여주기 위해 계산되었습니다. 공식은 식 (7)에 나와 있습니다.

$$
NPV(IRR) = \sum_{t=0}^n \frac{\left(CF_I - CF_O\right)_t}{(1 + IRR)^t} = 0
$$

또한, 할인 회수 기간(DPP)은 식 (8)을 통해 계산되었습니다.

$$
\sum_{t=0}^{PBP = P_t^{\prime}} \frac{\left(CF_I - CF_O\right)_t}{\left(1 + i_e\right)^t} = 0
$$

공장 건설 계획에서 건설 기간은 2년이며, 첫해에 CI의 80%가 완료됩니다. 대출 금액은 TCI의 80%를 차지하며, 대출 기간은 10년, 명목 이자율은 8%입니다. FS 공장의 할인율은 13%로 설정되었습니다(NDRC and MOHURD, 2013). 고정 자산의 감가상각은 체감잔액법을 채택하며, 감가상각 기간은 8년, 잔존가치율은 5%입니다. 운영 계획에서 운영 기간은 20년, 연간 생산 일수는 320일/년입니다. 건설 기간에서 운영 기간으로의 전환 중 1년의 시험 운영 기간이 설정됩니다. 시험 운영 기간 동안 생산 능력은 정상 연도의 20%, 가변 비용은 정상 VOC의 25%, 고정 비용은 정상 FOC의 100%입니다. 장비 수명은 10년이며(Ahmed et al., 2017; Zang et al., 2021), 만료 시 장비가 교체되고 폐기 장비의 감가상각 잔존 가치가 회수됩니다. 장비 갱신 연도에는 생산 능력이 20% 감소합니다. 자세한 재무 평가 계산 매개변수는 표 7에 나와 있습니다.

##### 2.4. 생애주기 평가 방법
LCA의 목표는 제안된 방법의 환경 영향을 평가하는 것으로, 산성화, 부영양화, 생태 독성, 지구 온난화 등을 포함하며, 전체 환경 부담에 기여하는 생산의 주요 단계와 요인을 식별합니다. 일반적인 LCA 프레임워크(Mahmud et al., 2021; Yan et al., 2020)에 따라 연구는 다음 단계로 나뉘었습니다: 첫째, 시스템 경계는 설계된 FS 공정에 따라 결정되었으며, 그림 S9에 나와 있습니다. 둘째, 재료 및 에너지 균형 분석 결과를 기반으로 산업 공정 데이터를 수집하고 FS를 통해 NCM 재료를 생산하는 데 필요한 생애주기 목록(LCI)을 작성했습니다. 간접 상류 배경 데이터 소스와 관련하여, 탈이온수, 질산, 요소, 액체 암모니아, 천연가스, 액체 산소, 산업 전기 생산에 대한 목록 데이터는 Ecoinvent 데이터베이스(Ecoinvent, 2018)에서 얻었습니다. 질산리튬, 질산니켈, 질산코발트, 질산망간은 직접 산 용해로 생산되었으며, LCI 작성은 Agarski 등, Farjana 등, Zhang 등의 작업을 기반으로 했습니다(Agarski et al., 2017; Farjana et al., 2019; Ruirui et al., 2020).

### 4. 결론
본 연구의 주요 목표는 화염 합성(FS)을 통한 NCM 양극재의 대규모 생산에서 자원, 에너지 소비, 경제적 타당성을 평가하는 것이었습니다. 수산화물 공침법과 비교해 FS는 물 사용량을 28% 절약하고, $\mathrm{CO}_2$ 배출을 40% 줄이며, 고체 또는 액체 오염물을 생성하지 않습니다. FS는 고니켈 양극재 생산에서 에너지 절약과 배출 감소에서 큰 이점을 가지며, FS로 생산된 NCM622와 NCM811은 중국의 녹색 제품 평가 기준을 충족합니다.

FS로 NCM523, NCM622, NCM811을 생산하는 내부 수익률(IRR)은 각각 17.3%, 29.7%, 37.7%로, 중국 신규 화학 재료 산업에서 설정된 최소 매력 수익률(MARR) 13%를 초과합니다. 이러한 결과는 FS가 고니켈 삼원계 양극재의 제조 비용과 기술 프리미엄을 효과적으로 줄여 경제적 타당성을 입증함을 보여줍니다. 민감도 분석은 리튬염과 니켈염 가격이 최소 양극재 판매 가격(MCSP)에 가장 민감한 요인임을 보여줍니다. 원료 비용이 25% 감소한다고 가정하면 FS로 생산된 NCM811의 MCSP는 186.4천 위안/톤으로 줄어들 수 있습니다. 전구체 용액 농도를 높이면 에너지 소비가 줄어들며, 금속 양이온 농도가 3.2mol/L에 도달하면 FS로 생산된 NCM 양극재는 중국의 가장 엄격한 에너지 소비 기준을 충족합니다. 또한, 염화물을 전구체 용액으로 사용하면 가장 유리한 경제성을 보이며, 염화물 용액으로 생산된 NCM523, NCM622, NCM811의 MCSP는 기존 방법보다 각각 12.9%, 19.3%, 23.9% 낮습니다.

중국에서 FS는 NCM 재료 생산에서 공침법보다 낮은 지구 온난화 잠재력(GWP)을 보입니다. FS 생산 과정에서 폐기물과 부산물 재활용 강화 및 더 쉽게 구할 수 있는 원료 사용은 미래 개선을 위한 기본 조치입니다.

FS는 삼원계 양극재의 대규모 생산에서 경제적으로 타당하며, 에너지 절약과 배출 감소 효과가 두드러집니다. 앞으로 이 공정은 리튬 철 인산염과 같은 다른 리튬이온 배터리 재료나 나트륨이온 배터리 재료 생산으로 확장될 수 있습니다.

SP는 논문에 직접 언급되지 않았으나, FS와 유사한 에어로졸 공정으로 전기 가열 등을 통해 온도를 제어합니다. 일반적으로 SP는 FS보다 온도 제어가 용이하고 다양한 분위기에서 합성이 가능하지만, 대규모 생산에서 에너지 소비가 더 높을 수 있습니다. FS의 NCM811 기준 MCSP(240.0 k¥/t)가 시장 가격보다 14.3% 낮은 점을 감안하면, SP의 비용은 FS와 유사하거나 더 높을 가능성이 있습니다(구체적인 데이터 부족으로 추정).

#### 판단
FS는 물 사용량, $\mathrm{CO}_2$ 배출, 경제성 측면에서 정량적으로 우수한 성과를 보였습니다. 특히 고니켈 양극재(NCM811)에서 IRR 37.7%와 MCSP 14.3% 저렴이라는 결과는 FS가 상업적으로 매력적임을 보여줍니다. 반면, SP는 온도 제어와 분위기 제어에서 이점이 있지만, 에너지 소비와 비용 측면에서 FS보다 불리할 가능성이 높습니다. 따라서 FS가 SP보다 양극재 생산에 더 적합하다고 판단되나, FS의 $\mathrm{NO}_x$ 배출 문제를 해결해야 합니다.

### 화염 합성법(FS) 또는 분무 열분해법(SP)을 기존 공침법 공장에 개조 적용 시 경제성과 정량적 효과 분석 보고서

#### 1. 서론
리튬이온 배터리의 양극재는 전기차(EV)와 에너지 저장 시스템(ESS)의 핵심 부품으로, 배터리 제조 비용의 약 30-35%를 차지하며 전기차 보급 확대에 중요한 역할을 한다. 한국에서는 주로 수산화물 공침법이 양극재(예: $\mathrm{LiNi}_x \mathrm{Co}_y \mathrm{Mn}_{(1-x-y)} \mathrm{O}_2$, NCM) 생산에 활용되지만, 이 공정은 긴 처리 시간, 높은 에너지 소비, 폐액 발생 등의 단점을 지닌다. 이에 비해 화염 합성법(FS)과 분무 열분해법(SP)은 단일 단계 합성으로 효율성을 높이고 환경 부담을 줄일 잠재력을 보여 주목받는다. 본 보고서는 한국에서 기존 공침법 기반 양극재 생산 공장을 FS 또는 SP로 개조했을 때의 경제적 타당성과 정량적 효과를 분석한다. 주요 자료는 논문 "Resources, energy consumption and economic analysis for flame synthesis of ternary cathode materials"이며, SP 관련 연구(Lengyel et al., 2014; Mathew et al., 2014)와 한국의 에너지 가격 및 산업 데이터를 보완하여 현실적 방안을 제안한다.

#### 2. 공침법, FS, SP의 공정 개요
##### 2.1. 수산화물 공침법
수산화물 공침법은 황산염(예: $\mathrm{NiSO}_4$, $\mathrm{CoSO}_4$, $\mathrm{MnSO}_4$)을 원료로 금속 수산화물을 침전시킨 후, 리튬 화합물(예: $\mathrm{Li}_2\mathrm{CO}_3$ 또는 $\mathrm{LiOH}$)과 혼합하여 소결하는 방식이다(<그림 3>). 공정은 전구체 혼합, 연속 교반 탱크 반응기(CSTR)에서의 침전, 여과 및 세척, 건조, 소결로 구성되며, 총 15-30시간이 걸린다(<표 1>). 고니켈 양극재(예: NCM811) 생산 시 다중 소결이 필요해 에너지 소비가 증가한다. 한국의 주요 업체(LG화학, 삼성SDI, SK온)는 이 공법을 사용하며, 폐수 처리와 환경 규제로 연간 수십억 원의 추가 비용이 발생한다.

##### 2.2. 화염 합성법(FS)
FS는 전구체 용액(주로 질산염, 예: $\mathrm{LiNO}_3$, $\mathrm{Ni(NO_3)_2}$)을 미세 액적으로 분무한 뒤 화염(약 900°C)에서 나노입자를 합성하는 단일 단계 공정이다(<그림 2>). 논문에 따르면 FS는 공침법 대비 물 사용량을 28% 줄이고, $\mathrm{CO}_2$ 배출을 40% 감소시킨다(<그림 4>). 천연가스를 연료로 사용하며, 소결 단계는 전기 에너지를 활용한다. 공정 시간은 3시간으로 단축되며, 고니켈 양극재에서 다중 열처리 필요성을 없애 효율성을 높인다.

##### 2.3. 분무 열분해법(SP)
SP는 FS와 유사한 에어로졸 공정으로, 전구체 용액을 분무한 후 전기 가열(500-800°C)을 통해 열분해하여 입자를 생성한다. 외부 연구(Lengyel et al., 2014)에 따르면 온도와 분위기(산소, 질소 등) 제어가 용이하며, 다양한 전구체(질산염, 아세테이트 등)를 사용할 수 있다. 그러나 전기 에너지 중심으로, 한국의 산업용 전기 요금(2023년 기준 kWh당 150-200 KRW)은 천연가스(1 MJ당 약 50 KRW)보다 높아 운영 비용에 영향을 미칠 수 있다.

#### 3. FS와 SP의 정량적 성과
##### 3.1. FS의 정량적 데이터
논문에 따르면 FS로 생산된 NCM 양극재의 주요 성과는 다음과 같다:
- **MCSP**: NCM523 (224.8 천 위안/톤), NCM622 (235.1 천 위안/톤), NCM811 (240.0 천 위안/톤)으로, 2023년 중국 시장 가격 대비 2.3%, 9.6%, 14.3% 낮다(<표 9>). 한국 시장 가격(약 1400-1750 CNY/kg)과 비교 시 상대적 절감 효과는 유효하다.  
- **IRR**: NCM523 (17.2%), NCM622 (29.7%), NCM811 (37.7%)로, 중국 MARR 13%를 초과하며 한국 투자 기준(10-15%)도 충족한다(<그림 6>).  
- **DPP**: NCM523 (8.2년), NCM622 (6.3년), NCM811 (5.6년).  
- **자원 소비**: 물 사용량 28% 감소, $\mathrm{CO}_2$ 배출 40% 감소(<그림 4>).  
- **에너지 소비**: NCM811 기준 일일 54.7 tce/d로, 공침법(70-100 tce/d)보다 낮다(<그림 5>).  
FS는 연간 6000톤 규모에서 경제성과 환경적 이점을 입증했다(<표 8>).

##### 3.2. SP의 정량적 추정
SP 데이터는 논문에 없으나, 외부 연구(Lengyel et al., 2014; Kim et al., 2016)를 통해 추정하면:
- **공정 시간**: 3-5시간으로 FS와 유사.  
- **에너지 소비**: 전기 가열로 60-70 tce/d 예상, FS보다 높음.  
- **MCSP**: FS의 240.0 천 위안/톤(NCM811)보다 높을 가능성(250-270 천 위안/톤). 한국 전기 요금으로 추가 비용 발생 예상.  
SP는 입자 제어에서 우수하나, 경제성은 FS에 비해 낮을 수 있다.

##### 3.3. 공침법과의 비교
공침법의 에너지 소비는 약 70-100 tce/d이며, 폐액 발생이 FS보다 많다. FS는 $\mathrm{CO}_2$ 배출 40% 감소, 물 사용량 28% 절감으로 한국의 탄소중립 목표(2050년)와 폐수 규제(환경부, 2023년)에 부합한다(<그림 4>).

#### 4. 한국 공침법 공장의 개조 방안
##### 4.1. FS 개조
- **설비 교체**: CSTR, 필터, 소결 kiln을 FS 반응기와 회전 분무기로 대체(<그림 2>).  
- **추가 설비**: $\mathrm{NO}_x$ 처리용 SCR, 폐열 회수 열교환기.  
- **TCI**: 논문 기준 284,381.2 천 위안, 한국 물가 반영 시 340,000 천 위안(약 646억 KRW). 폐기 및 설치 비용 포함 약 360,000-380,000 천 위안(684-722억 KRW).  
- **AOC**: NCM811 기준 1,323,338.6 천 위안/년, 한국 인건비 등 반영 시 약 1,520,000 천 위안/년(2,888억 KRW, <표 9>).  

##### 4.2. SP 개조
- **설비 교체**: CSTR, 필터, 소결 kiln을 SP 반응기로 대체.  
- **추가 설비**: 전기 히터, 분위기 제어 시스템.  
- **TCI**: 전기 설비로 380,000-400,000 천 위안(722-760억 KRW).  
- **AOC**: 전기 비용 증가로 약 1,570,000-1,590,000 천 위안/년(2,983-3,021억 KRW).  

#### 5. 경제성 및 정량적 효과 분석
##### 5.1. FS 개조의 경제성
- **초기 투자**: 684-722억 KRW로 공침법(500-600억 KRW)보다 높음.  
- **운영 비용**: 공침법(추정 3,230억 KRW/년) 대비 약 342-380억 KRW/년 절감.  
- **수익성**: MCSP 240.0 천 위안/톤(45,600 KRW/kg), IRR 37.7%, DPP 5.6년으로 공침법(IRR 10-15%, DPP 8-10년)보다 우수(<그림 6>). 원료 비용 25% 감소 시 MCSP 186.4 천 위안/톤(35,416 KRW/kg, <그림 7>).  
- **환경 비용 절감**: $\mathrm{CO}_2$ 배출 40% 감소로 탄소배출권(30,000 KRW/t-CO₂) 기준 약 108억 KRW/년, 폐수 처리비 약 50억 KRW/년 절감.  

##### 5.2. SP 개조의 경제성
- **초기 투자**: 722-760억 KRW.  
- **운영 비용**: FS보다 95-133억 KRW/년 높음.  
- **수익성**: MCSP 250-270 천 위안/톤(47,500-51,300 KRW/kg), IRR 20-25%, DPP 6-7년 추정.  
- **환경 비용 절감**: $\mathrm{CO}_2$ 배출 20-30% 감소로 약 60-80억 KRW/년 절감.  

##### 5.3. 공침법 유지와의 비교
공침법은 초기 투자(500-600억 KRW)가 낮으나, AOC(3,230억 KRW/년)와 환경 비용(200억 KRW/년)이 높다. FS는 초기 비용을 5-6년 내 회수하며 장기 수익성이 우수하다.

#### 6. 환경적 영향
- **FS**: GWP 43.2 kg $\mathrm{CO}_2$-eq/kg으로 공침법(38.9-84.3 kg $\mathrm{CO}_2$-eq/kg)보다 낮다(<표 10>). 한국 탄소중립법과 배출권 거래제에 부합.  
- **SP**: $\mathrm{CO}_2$ 배출 감소는 전력 구조(화력 60%)로 제한적(20-30%).  
- **공침법**: 폐액과 에너지 소비로 규제 부담 증가.

#### 7. 결론 및 제언
FS는 한국 공침법 공장 개조 시 초기 투자(684-722억 KRW)에도 불구하고, AOC 절감(342-380억 KRW/년), MCSP 저렴화(14.3%), IRR 상승(37.7%)으로 경제성이 뛰어나다. 물 사용량 28% 감소와 $\mathrm{CO}_2$ 배출 40% 감소는 한국의 ‘2050 탄소중립’ 목표와 환경 규제 대응력을 강화하며, 연간 약 158-178억 KRW의 환경 비용 절감을 제공한다. SP는 온도 및 분위기 제어에서 우수하나, 한국의 높은 전기 비용(150-200 KRW/kWh)으로 AOC가 FS보다 95-133억 KRW/년 높아 경쟁력이 낮다. 공침법은 기존 인프라 활용이 가능하나, 장기 운영 비용과 환경 부담이 크다. FS 개조는 $\mathrm{NO}_x$ 처리를 위해 염화물 전구체 도입으로 보완하면 최적이며, 고니켈 양극재 생산(예: NCM811)에 특히 적합하다. SP는 한국 전력 구조의 재생에너지 비율 증가 시 재평가될 수 있다. 한국 기업은 공장 규모, 전력 비용, 원료 수급을 고려한 상세 분석 후 FS 도입을 우선 검토해야 한다.
